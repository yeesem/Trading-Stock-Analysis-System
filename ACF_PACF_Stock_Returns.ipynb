```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/yeesem/Time-Series-Data-Analysis/main/sp500sub.csv",index_col = 'Date',parse_dates = True)
```

```python
df.head()
```

```python
goog = df[df['Name'] == 'GOOG'][['Close']].copy()
```

```python
def log_diff(price):
  log_price = np.log(price)
  return log_price .diff()
```

```python
goog['LogPriceDiff'] = log_diff(goog['Close'])
```

```python
goog.head()
```

```python
goog['LogPriceDiff'] = log_diff(goog['Close'])
```

```python
goog.head()
```

```python
p = 0
fig,ax = plt.subplots(figsize = (10,5))
plot_pacf(goog['LogPriceDiff'].dropna() ,ax = ax);
```

```python
#q = 0
fig,ax = plt.subplots(figsize = (10,5))
plot_acf(goog['LogPriceDiff'].dropna(),ax =ax);
```
